```python
# i chose to use the following kod to identify 
import csv
def reading_csvfile():
    pokemons = []

    with open("datapoints.csv") as csvfile:
        reader = csv.DictReader(csvfile)
        
        for row in reader:
            width_label = '(width (cm)'
            height_label = ' height (cm)'
            label_label  = ' label (0-pichu 1-pikachu))'
            
            width = float(row[width_label])
            height = float(row[height_label])
            label = int(row[label_label])

            pokemons.append([width, height, label])

    return pokemons

 
```

```python
# läs från filen, skippa första reden, sedan för varje rad, skippa 4 första tecknen, ta bort ) på slutet, och splitta på komma (,) och konverta varje sträng till float.
import math
def read_from_file():
    


    test_point_list = []
    with open("testpoints.txt", "r") as file:
       # distance_result_list = []

        for row in file.readlines()[1:]: 
            
        
            features = row[4:-2].split(", ")
            width = float(features[0])
            height = float(features[1])

            test_point_list.append({"width": width, "height": height})
    
    return test_point_list







def distance_classification(test_point_list, pokemons):

    sorted_distance_list = []

    for testpoint in test_point_list:
        x1 = testpoint['width']
        y1 = testpoint['height']
        
        distance_list = []

        for pokemon in pokemons:
            y2 = pokemon[1]
            x2 = pokemon[0]
            label = pokemon[2]
            
            distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
            distance_label = [distance, label]

            distance_list.append(distance_label)

        sorted_list = sorted(distance_list)
        sorted_distance_list.append(sorted_list)

    for item in sorted_distance_list:
        
        d1 = item[0][1]

        if d1 == 0:
            print("item is pichu")

        else: 
            print("item is pikachu")
        

    
```

```python
import matplotlib.pyplot as plt 

def plot_points(pokemons):
    

    pika_widths = [pokemon[0] for pokemon in pokemons if pokemon[2]== 1]
    pika_heights = [pokemon[1] for pokemon in pokemons if pokemon[2] == 1]
    plt.scatter(pika_widths, pika_heights, color = "red")

    pichu_widths = [pokemon[0] for pokemon in pokemons if pokemon[2]== 0]
    pichu_heights = [pokemon[1] for pokemon in pokemons if pokemon[2] == 0]
    plt.scatter(pichu_widths, pichu_heights, color = "blue")

    plt.show()


```

```python
def input_testpoint(pokemons):
    while True:
        try:
            width_input = float(input("Enter width: "))
            height_input = float(input("enter height: "))
        
        
        except ValueError:
            print("Value cannot be string")
            continue
        if width_input < 0 or height_input < 0:
            print ("value must be positive")
        else: 
            test_points = {"width": width_input, "height": height_input}
            distance_classification([test_points], pokemons)
            return


        

    
        
```

```python
def distance_classification_10(test_point_list, pokemons):
    
    sorted_distance_list = []

    for testpoint in test_point_list:
        x1 = testpoint['width']
        y1 = testpoint['height']
        
        distance_list = []

        for pokemon in pokemons:
            y2 = pokemon[1]
            x2 = pokemon[0]
            label = pokemon[2]
            
            distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
            distance_label = [distance, label]

            distance_list.append(distance_label)

        sorted_list = sorted(distance_list)
        sorted_distance_list.append(sorted_list[:10])
     
    
    for item in sorted_distance_list:
        distance_pikachu = 0
        distance_pichu = 0
        vote_for_pikachu = 0
        vote_for_pichu = 0
        for sorted_distance in item:


            if sorted_distance[1] == 1:
                distance_pikachu += sorted_distance[0]
                vote_for_pikachu += sorted_distance[1]

            else: 
                distance_pichu += sorted_distance[0]
                vote_for_pichu += 1 
        
        if vote_for_pichu > 5:
            print("its pichu")
        
        if vote_for_pikachu > 5:
            print("its pikachu")
        
        if vote_for_pikachu == 5:
            
            if distance_pichu > distance_pikachu:
                print("it is pikachu")

            else:

                print("Its Pichu")

        

        


        
        
        
        
    
        
```
